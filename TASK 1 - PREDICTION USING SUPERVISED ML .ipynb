The given objective is to predict the percentage of an student based on the no. of study hours. We will use Simple Linear Regression algorithm along which involves just 2 variables and find out, What will be predicted score if a student studies for 9.25 hrs/ day?

```python
#Importing all the required libraries needed for this action
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
#Importing the data
dt=pd.read_csv("http://bit.ly/w-data")
dt
```

```python
# Check if there any null value in the Dataset, If false means no value is missing
dt.isnull == True
```

```python
dt.describe()
```

# GRAPH PLOTTING/ DATA VISUALIZING

```python
#Plotting Grid graph to show its scales
dt.plot(kind = 'scatter', x = 'Hours', y = 'Scores', style = '-')
plt.title("Hours vs Percentage")
plt.xlabel('Hours studied')
plt.ylabel('Scores obtained')
plt.grid()
plt.show()
```

```python
#Plotting Bar Graph for the same
dt.plot(kind='bar',figsize=(8,5))
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

# Now we will Prepare, Split and Test the data.

```python
#Prepare, Splitting the training and testing it
x = dt.iloc[:,:-1].values
y = dt.iloc[:,1].values

#Splitting it into two 
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 0)
```

# MODEL AND ALGORITHM TRAINING

```python
linearRegressor= LinearRegression()
linearRegressor.fit(train_x, train_y)
y_predict=linearRegressor.predict(train_x)
```

```python
regression = LinearRegression()
regression.fit(train_x, train_y)
print("Training is successfully completed")
```

# REGRESSION PLOT

We'll derive here regression coefficient and regression intercept to plot regression line.

```python
line = linearRegressor.coef_*x+linearRegressor.intercept_
```

```python
plt.scatter(x, y)
plt.plot(x, line, color = 'red')
plt.title('Hours vs Percentage')
plt.show()
```

# MODEL ACCURACY

```python
print('Test score = ')
print(linearRegressor.score(val_x, val_y))
print('Training score = ')
print(linearRegressor.score(train_x, train_y))
```

# PREDICTING PERCENTAGE OF MARKS

```python
pred_y = regression.predict(val_x)
prediction = pd.DataFrame({'Hours': [i[0] for i in val_x], 'Predicted Marks': [k for k in pred_y]})
prediction
```

# COMPARING PREDICTED V/S ACTUAL MARKS

```python
compare_scores = pd.DataFrame({'Actual Marks': val_y, 'Predicted Marks': pred_y})
compare_scores
```

Visual Comparison of Predicted marks and Actual marks

```python
plt.scatter(x=val_x, y=val_y, color='blue')
plt.plot(val_x, pred_y, color='Black')
plt.title('Actual vs Predicted', size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.grid()
plt.show()
```

# EFFICIENCY OF MODEL

```python
#Calculating accuracy of the model
from sklearn import metrics
print('Mean absolute error: ',mean_absolute_error(val_y,pred_y))
```

# SUB-TASK SOLUTION

So sub-task 2, i.e. What will be the predicted score of a student if he/she studies for 9.25 hrs/ day?

```python

hours = [9.25]
answer = regression.predict([hours])
print("Score = {}".format(round(answer[0],3)))
```
